<div dir="rtl" align="justify">
    <ul>
        <li>برای تمام شکست های ممکن ضریب جینی محاسبه میشود. سپس شکستی که <u>کمترین</u> مقدار gini را دارد، انتخاب میشود.</li>

        <li>بد ترین حالت gini زمانی رخ میدهد که تعداد رکورد های هر دسته در آن گره با هم برابر باشند. در این حالت gini برابر است با 0.5</li>
        <li>برای ویژگی های پیوسته: <br>
داده ها را طبقه بندی میکنیم. سپس بر روی نماینده هر طبقه، مقدار ضریب gini را محاسبه میکنیم. بر مبنای نماینده با کم ترین ضریب gini، مقادیر را به دو شاخه تقسیم میکنیم.</li>

</div>

```python
import pandas as pd
```

```python
dataset =  pd.read_csv('/mnt/f/internetlogit.csv')
```

```python
dataset.head
```

```python
X=dataset.iloc[:,0:5].values
y=dataset.iloc[:,5].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.3, random_state=0)
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn import tree, metrics
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
clf = clf.fit(X_train, y_train)
```

```python
text_representation = tree.export_text(clf)
print(text_representation)
```

```python
from sklearn.tree import export_graphviz
from six import StringIO
from IPython.display import Image  
import pydotplus
feature_cols = ['age', 'gender', 'webpages', 'videohours' 'income','usage']
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,feature_names = feature_cols,class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('diabetes.png')
Image(graph.create_png())
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
